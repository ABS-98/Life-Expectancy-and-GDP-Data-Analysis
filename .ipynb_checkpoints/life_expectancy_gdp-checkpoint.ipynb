# Import Python Modules

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

# Loading The Data

```python
data = pd.read_csv('all_data.csv')
data.head(10)
```

# Inspecting The Data

```python
data.info()
data.describe(include = 'all')
```

*There seem to be no missing data or outliers in the dataset, so no data cleaning is needed.*

# Exploring the Data
First let's check the unique values we have in these columns.

```python
print(data.Country.unique())
print(data.Year.unique())
```

*The dataset has six unique countries which are (Chile - Chine - Germany - Mexico - USA - Zimbabwe), and the years the data were collected ranges from 2000 to 2015.*

*The country `United States of America` name is too long and might cause problems in the plots later, let's replace it with the appreviation `USA`*

```python
data.Country.replace('United States of America', 'USA', inplace = True)
```

```python
print(data.Country.unique())
```

#### Cleaning columns names
The column `Life expectancy at birth (years)` is not practical, let's change it into something more suitable.

```python
data.rename({'Life expectancy at birth (years)': 'life_expectancy', 'Country': 'country', 'Year': 'year', 'GDP': 'gdp'},
            axis = 'columns', inplace = True)
data.head()
```

### Exploratory Data Analysis (EDA)


```python
plt.figure(figsize = (5,5))
sns.displot(data.gdp, rug = True)
plt.xlabel('GDP in Trillions of US Dollars')
plt.title('GDP Distribution')
plt.show()
plt.clf()
```

*The distribution of the `GDP` is highly right skewed where most of the values are on the left hand side.*

```python
plt.figure(figsize = (5,5))
sns.displot(data.life_expectancy, rug = True)
plt.xlabel('Life expectancy at birth (years)')
plt.title('Life Expectancy Distribution')
plt.show()
plt.clf()
```

*The `Life expectancy at birth (years)` distribution is left skewed, the distribution also shows different modes or small groupings.*

*The previous plots just show the distribution of the data, now let's break them down by country*

```python
data_means = data.drop('year', axis = 1).groupby('country').mean().reset_index()
data_means
```

```python
plt.figure(figsize = (10, 5))
ax1 = plt.subplot(1, 2, 1)
sns.barplot(x = 'country', y = 'gdp', data = data_means)
ax1.set_xticklabels(data_means.country, rotation = 30)
plt.ylabel('GDP in Trillions of US Dollars')
plt.xlabel('Country')
plt.title('Mean GDP for each Country')
ax2 = plt.subplot(1, 2, 2)
sns.barplot(x = 'country', y = 'life_expectancy', data = data_means)
ax2.set_xticklabels(data_means.country, rotation = 30)
plt.xlabel('Country')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.title('Mean Life Expectancy for each Country')
plt.subplots_adjust(wspace = 0.35)
plt.show()
plt.clf()
```

* *The GDP plot shows that `USA` has a much higher average GDP compared to the rest of the countries, `Zimbabwe` is not even visible and `Chile` is barely visible, this explains the skewness in the GDP distribution*
* *The life expectancy plot shows that all countries except `Zimbabwe` have mid to high average life expectancy in the range of 70-80s which explains the skewness in the distribution plot.*

```python
plt.figure(figsize = (10, 6))
ax1 = plt.subplot(1, 2, 1)
sns.boxplot(x = 'country', y = 'gdp', data = data)
sns.stripplot(x = 'country', y = 'gdp', data = data, size = 5, alpha = 0.5)
ax1.set_xticklabels(data_means.country, rotation = 30)
plt.ylabel('GDP in Trillions of US Dollars')
plt.xlabel('Country')
plt.title('GDP for each Country')
ax2 = plt.subplot(1, 2, 2)
sns.boxplot(x = 'country', y = 'life_expectancy', data = data)
sns.stripplot(x = 'country', y = 'life_expectancy', data = data, size = 5, alpha = 0.5)
ax2.set_xticklabels(data_means.country, rotation = 30)
plt.xlabel('Country')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.title('Life Expectancy for each Country')
plt.subplots_adjust(wspace = 0.25)
plt.show()
plt.clf()
```

*The box plot combined with a strip plot explain the mean plots we had previously shown, if we look at China's `GDP` the box plot shows the wide range of `GDP China` has, and the strip plot explains why the average `GDP` for `China` is at the lower end of the distribution, which is due to the presence of a big cluster of data at the lower end of the distribution, same thing is showing for `USA` but on the higher end of the distribution.
This also explains the low average `Life Expectancy at Birth (Years)` for `Zimbabwe` due to the presence of a cluster of data at the lower end of the range despite the wide range the country has.*

### Time Series Analysis

```python
plt.figure(figsize = (8,5))
ax = plt.subplot()
sns.lineplot(x= 'year', y= 'gdp', hue = 'country', data = data)
plt.legend(loc = 'center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.xlabel('Year')
plt.ylabel('GDP in Trillions of US Dollars')
ax.set_xticks(data.year.unique())
ax.set_xticklabels(data.year.unique(), rotation = 30)
plt.show()
plt.clf()
```

*The plot shows the substantial growth in `GDP` the `USA` and `China` had over time, `USA` went from 1 trillion to 1.75 trillions, while `China` went from less than quarter trillion dollars to one trillion dollars. The rest of the countries however did not have growth in such magnitude.*

```python
gdp_graph = sns.FacetGrid(data = data, col= 'country', hue = 'country', col_wrap = 3, sharey = False)
gdp_graph = gdp_graph.map(sns.lineplot, 'year', 'gdp').add_legend().set_axis_labels('Year', 'GDP in Trillions of US Dollars')
plt.show()
plt.clf()
```

*In the figure above we plotted each country individually to better visualize the growth of each country. All countries have seen increases over the years even if it's not as big as the growth `China and USA` have seen.*

```python
plt.figure(figsize= (8, 5))
ax = plt.subplot()
sns.lineplot(x = 'year', y = 'life_expectancy', hue = 'country', data = data)
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5), ncol = 1)
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (Years)')
ax.set_xticks(data.year.unique())
ax.set_xticklabels(data.year.unique(), rotation = 30)
plt.show()
plt.clf()
```

*The plot shows that all countries had an increase in the `Life Expectancy`. However, `Zimbabwe` had the greatest increase after a dip in 2004.*

```python
life_expectancy_graph = sns.FacetGrid(data = data, col = 'country', hue = 'country', col_wrap = 3, sharey = False)
life_expectancy_graph = life_expectancy_graph.map(sns.lineplot, 'year', 'life_expectancy').add_legend()\
.set_axis_labels('Year', 'Life Expectancy at Birth (Years)')
plt.show()
plt.clf()
```

*The breakdown of individual countries with each their own scale shows a better analysis of the graphs, now it's apparent that `Mexico and Chile` both had dips arount the same time (2010) which can be looked into further.*

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'life_expectancy', y = 'gdp', hue = 'country', data = data)
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5), ncol = 1)
plt.xlabel('Life Expectancy at Birth (Years)')
plt.ylabel('GDP in Trillions of US Dollars')
plt.show()
plt.clf()
```

*Most countries show a positive correlation between `Life Expectancy and GDP`. However, `Zimbabwe's` GDP looks flat while its Life Expectancy increases, let's take a closer look on each country individually for better visualization.*

```python
gdp_lexpec = sns.FacetGrid(data = data, col = 'country', hue = 'country', col_wrap = 3, sharey = False, sharex = False)
gdp_lexpec = gdp_lexpec.map(sns.scatterplot, 'life_expectancy', 'gdp').add_legend()\
.set_axis_labels( 'Life Expectancy at Birth (Years)', 'GDP in Trillions of US Dollars')
plt.show()
plt.clf()
```

*Now it's apparent that most countries indeed have a positive correlation between `GDP and Life Expectancy` including `Zimbabwe` which looked like it has no correlation in the previous graph, `China` on the other hand has an exponential growth and `Chile` looks a bit logarethmic. In General, all countries have a positive correlation.*
